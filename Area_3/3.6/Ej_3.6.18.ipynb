```python
import pandas as pd

route = 'C:/Users/maite/OneDrive/Escritorio/Curso-NTT/Area_3/3.6/archivos/netflix_movies_and_tv_shows_sample_dataset_sample.csv'
df = pd.read_csv(route, sep=',')
df
```

```python
# Features
df.info()
```

```python
# Comprobar nulos
df.isnull().sum()
```

```python
# Como 'formattedDuration', 'director', 'creator', 'audio' y 'subtitle' tienen más del 50% de valores nulos
# se eliminan
columnas = ['formattedDuration', 'director', 'creator', 'audio', 'subtitle']
df[columnas] = df[columnas].fillna(0)
df.drop(columns=columnas, inplace=True)

df
```

```python
# En las columnas 'actors', 'numberOfSeasons' y 'seasonStartDate' no se pueden imputar los registros nulos
# ya que no son de tipo numérico. Por lo que se eliminarán estos registros nulos.
df = df[(~df.actors.isnull())]
df = df[(~df.numberOfSeasons.isnull())]
df = df[(~df.seasonStartDate.isnull())]

df.isnull().sum()
```

```python
# Reseteo índices
df = df.reset_index()
df
```

```python
# Guardar el DataFrame limpio en un nuevo csv
df.to_csv('./archivos/clean_netflix_movies_and_tv_shows_sample_dataset_sample.csv', sep = ',')
```
